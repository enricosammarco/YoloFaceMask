```
import xml.etree.ElementTree as ET
import os

f = open("face_mask_train_corrette.txt", "w")

for i in os.listdir("train/"):
  if "xml" in i:
    tree = ET.parse("train/"+i)
    root = tree.getroot()
    filename = "train/"+root[1].text.replace("xml","jpg")
    riga = filename
    for c in root:
      if len(c)>0 and c[0].tag == "name":
        riga = riga + " "
        classe = c[0].text
        for child in c:
          if child.tag == "bndbox":
            bb = child
            sbb = bb[0].text+","+bb[1].text+","+bb[2].text+","+bb[3].text+","
            print(sbb)
        if classe == "face":
          riga = riga + sbb+"0"
        elif classe ==  "face_mask" or classe =="face_nask":
          riga = riga + sbb+"1"
    f.write(riga+"\n")
f.close()

f = open("face_mask_train_corrette.txt", "r")
print(len(f.readlines()))
f.close()
```
